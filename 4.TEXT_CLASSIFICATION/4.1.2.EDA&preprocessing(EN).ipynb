### 데이터 전처리

데이터 분석과정을 바탕으로 데이터를 모델에 적용시키기 위해 전처리 과정을 진행한다.

```python
import re
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
from tensorflow.python.keras.preprocessing.text import Tokenizer
```

#### 1. 데이터 하나를 자세히 확인

```python
DATA_IN_PATH = './data_in/'

train_data = pd.read_csv( DATA_IN_PATH + 'labeledTrainData.tsv', header = 0, delimiter = '\t', quoting = 3)
print(train_data['review'][0])
```

#### 2. 전처리 방향 결정

```python
review = train_data['review'][0] # 리뷰 중 하나를 가져온다.
# review_text = BeautifulSoup(review,"html5lib").get_text() # HTML 태그 제거
review_text = BeautifulSoup(review,"html.parser").get_text() # HTML 태그 제거
review_text = re.sub("[^a-zA-Z]", " ", review_text ) # 영어 문자를 제외한 나머지는 모두 공백으로 바꾼다.
```

```python
print(review_text)
```

#### 2. (1) 불용어(stopword) 삭제: 문장에서 자주 출현하나 전체적인 의미에 큰 영향을 주지 않는 단어(노이즈로 간주)

```python
stop_words = set(stopwords.words('english')) # 영어 불용어들의 set을 만든다.

review_text = review_text.lower()
words = review_text.split() # 소문자 변환 후 단어마다 나눠서 단어 리스트로 만든다.
words = [w for w in words if not w in stop_words] # 불용어 제거한 리스트를 만든다
```

```python
print(words)
```

```python
clean_review = ' '.join(words) # 단어 리스트들을 다시 하나의 글로 합친다.
print(clean_review)
```

#### 3. 전처리 함수 정의

```python
def preprocessing( review, remove_stopwords = False ): 
    # 불용어 제거는 옵션으로 선택 가능하다.
    
    # 1. HTML 태그 제거
#     review_text = BeautifulSoup(review, "html5lib").get_text()
    review_text = BeautifulSoup(review,"html.parser").get_text() # HTML 태그 제거


    # 2. 영어가 아닌 특수문자들을 공백(" ")으로 바꾸기
    review_text = re.sub("[^a-zA-Z]", " ", review_text)

    # 3. 대문자들을 소문자로 바꾸고 공백단위로 텍스트들 나눠서 리스트로 만든다.
    words = review_text.lower().split()

    if remove_stopwords: 
        # 4. 불용어들을 제거
    
        #영어에 관련된 불용어 불러오기
        stops = set(stopwords.words("english"))
        # 불용어가 아닌 단어들로 이루어진 새로운 리스트 생성
        words = [w for w in words if not w in stops]
#         # 5. 단어 리스트를 공백을 넣어서 하나의 글로 합친다.
#         clean_review = ' '.join(words)

#     else: # 불용어 제거하지 않을 때
#         clean_review = ' '.join(words)
    
    # 5. 단어 리스트를 공백을 넣어서 하나의 글로 합친다.
    clean_review = ' '.join(words)

    return clean_review
```

```python
clean_train_reviews = []
for review in train_data['review']:
    clean_train_reviews.append(preprocessing(review, remove_stopwords = True))

# 전처리한 데이터 출력
clean_train_reviews[0]
```

```python
# (전처리 후) 많이 사용되는 단어 확인
from wordcloud import WordCloud
cloud = WordCloud(width=600, height=500).generate(" ".join(clean_train_reviews))
plt.figure(figsize=(20, 15))
plt.imshow(cloud)
plt.axis('off')
```

#### 4. 벡터화

    - 입력값의 길이를 동일하게 변환 (일정 길이로 자르거나 특정값을 채우는 패딩)
    - 단, 벡터가 아닌 텍스트로 구성되어야하는 경우가 있음 (텍스트 데이터 저장 in pd)

#### 4.1 원본 데이터가 벡터화 되는 과정

1. 원본
2. 인덱스 변환(단어 사전 만들기)
3. 사이즈 패딩
4. 벡터화

```python
clean_train_df = pd.DataFrame({'review': clean_train_reviews, 'sentiment': train_data['sentiment']})
```

A. 정제된 데이터를 벡터로 변환

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_train_reviews)  # 각 단어를 인덱스로 만듬
text_sequences = tokenizer.texts_to_sequences(clean_train_reviews) # 텍스트 데이터를 인덱스로 변환
```

```python
print(text_sequences[0])
```

B. 단어 사전

```python
word_vocab = tokenizer.word_index
print(word_vocab)
```

```python
print("전체 단어 개수: ", len(word_vocab) + 1)
```

```python
tokenizer.word_index
```

```python
data_configs = {}  # 단어사전과 갯수 저장

data_configs['vocab'] = word_vocab
data_configs['vocab_size'] = len(word_vocab)
```

C. 특정길이로 정하여 자르거나 패딩하는 작업

```python
np.median(clean_train_df['review'].apply(lambda x:len(x.split(' '))))  # 정제 후
```

```python
MAX_SEQUENCE_LENGTH = 174  # (정제전) 중앙값으로 길이 정함 -> Feature를 선정하는 것은 먼저 탐색하면서 정함, 하기 나름...ㅎ

train_inputs = pad_sequences(text_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post')

print('Shape of train data: ', train_inputs.shape)
```

```python
train_labels = np.array(train_data['sentiment'])
print('Shape of label tensor:', train_labels.shape)
```

#### 5. 데이터 저장

    - 정제된 텍스트 데이터
    - 벡터화한 데이터
    - 정답 라벨
    - 데이터 정보(단어사전, 전체 단어 갯수)

```python
TRAIN_INPUT_DATA = 'train_input.npy'  # numpy배열로 저장 -> index로 만들어야함
TRAIN_LABEL_DATA = 'train_label.npy'
TRAIN_CLEAN_DATA = 'train_clean.csv'  # input과 label이 함께 있음, 전처리 후 문자열(char)로 가지고 있음 -> 벡터화
DATA_CONFIGS = 'data_configs.json'

import os
# 저장하는 디렉토리가 존재하지 않으면 생성
if not os.path.exists(DATA_IN_PATH):
    os.makedirs(DATA_IN_PATH)
```

```python
# 전처리 된 데이터를 넘파이 형태로 저장
np.save(open(DATA_IN_PATH + TRAIN_INPUT_DATA, 'wb'), train_inputs)
np.save(open(DATA_IN_PATH + TRAIN_LABEL_DATA, 'wb'), train_labels)

# 정제된 텍스트를 csv 형태로 저장
clean_train_df.to_csv(DATA_IN_PATH + TRAIN_CLEAN_DATA, index = False)

# 데이터 사전을 json 형태로 저장
json.dump(data_configs, open(DATA_IN_PATH + DATA_CONFIGS, 'w'), ensure_ascii=False)
```

```python
# test 데이터에도 train 데이터와 같은 전처리를 적용함.
test_data = pd.read_csv(DATA_IN_PATH + "testData.tsv", header=0, delimiter="\t", quoting=3)

clean_test_reviews = []
for review in test_data['review']:
    clean_test_reviews.append(preprocessing(review, remove_stopwords = True))

clean_test_df = pd.DataFrame({'review': clean_test_reviews, 'id': test_data['id']})
test_id = np.array(test_data['id'])

# tokenizer.fit_on_text(clean_test_reviews)    # 데이터 사전은 train으로 만든 사전을 동일하게 이용

text_sequences = tokenizer.texts_to_sequences(clean_test_reviews)
test_inputs = pad_sequences(text_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post')
```

```python
TEST_INPUT_DATA = 'test_input.npy'
TEST_CLEAN_DATA = 'test_clean.csv'
TEST_ID_DATA = 'test_id.npy'

np.save(open(DATA_IN_PATH + TEST_INPUT_DATA, 'wb'), test_inputs)
np.save(open(DATA_IN_PATH + TEST_ID_DATA, 'wb'), test_id)
clean_test_df.to_csv(DATA_IN_PATH + TEST_CLEAN_DATA, index = False)
```
